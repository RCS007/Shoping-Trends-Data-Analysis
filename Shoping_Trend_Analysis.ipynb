### Exploratory Data Analysis

#### 1. Distribution of Customer Ages.

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.grid(False)
plt.figtext(0.5, -0.1, 'This histogram shows the distribution of customer ages, indicating the most common age groups.', ha='center', fontsize=12)
plt.show()
```

####  2. Average purchase amount across product categories

```python
avg_purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].mean().reset_index()
sns.barplot(data=avg_purchase_by_category, x='Category', y='Purchase Amount (USD)', palette='viridis')
plt.title('Average Purchase Amount by Category')
plt.xlabel('Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.figtext(0.5, -0.1, 'This Barplot shows that average purchase amount vary across differnt categories such as Clothing, Accesories, Footware, and Outerware.', ha='center', fontsize=12)
plt.show()
```

####  3. Gender with the highest number of purchases

```python
gender_purchase = df.groupby('Gender')['Purchase Amount (USD)'].sum().reset_index()
sns.barplot(data=gender_purchase, x='Gender', y='Purchase Amount (USD)', palette='pastel')
plt.title('Total Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Purchase Amount (USD)')
plt.figtext(0.5, -0.1, 'This Barplot shows that Male customers are more as compared to Female customers also they are spending more money or we can say that they have high purchasing power.', ha='center', fontsize=12)
plt.show()
```

#### 4. Commonly purchased items in each category

```python
most_common_items = df.groupby(['Category', 'Item Purchased']).size().reset_index(name='Count')
most_common_items = most_common_items.sort_values(by=['Category', 'Count'], ascending=[True, False])
plt.figure(figsize=(10, 6))
sns.barplot(data=most_common_items, y='Item Purchased', x='Count', hue='Category', dodge=False)
plt.title('Most Commonly Purchased Items by Category')
plt.xlabel('Count')
plt.ylabel('Item Purchased')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### This Barplot shows commonly purchased items in category as follows:-
##### 1. Clothing Category
#####    *   Blouse, Pants, Shirt, and Dress.
##### 2. Accesories Category
#####    *   Jewelry, Belt, Sunglasses.
##### 3. Footware
#####    *   Sandals, Shoes.
##### 4. Outerware
#####    *   Jacket, Coat.

#### 5. Purchase frequency by season

```python
purchase_by_season = df['Season'].value_counts()
purchase_by_season.plot(kind='bar', color='lightgreen', figsize=(8, 6))
plt.title('Purchase Frequency by Season')
plt.xlabel('Season')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.figtext(0.5, -0.1, 'This bar chart shows the frequency of purchases in each season, indicating trends in customer buying behavior.', ha='center', fontsize=12)
plt.show()
```

#### 6. Average rating by product category

```python
avg_rating_by_category = df.groupby('Category')['Review Rating'].mean()
avg_rating_by_category.plot(kind='bar', color='orange', figsize=(8, 6))
plt.title('Average Review Rating by Product Category')
plt.xlabel('Category')
plt.ylabel('Average Review Rating')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.figtext(0.5, -0.1, 'This chart shows the average review rating for each product category, indicating customer satisfaction.', ha='center', fontsize=12)
plt.show()
```

#### 7. Differences in purchase behavior between subscribed and non-subscribed customers

```python
subscription_counts = df['Subscription Status'].value_counts()
plt.figure(figsize=(8, 8))
subscription_counts.plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'skyblue'])
plt.title('Breakdown of Customers by Subscription Status')
plt.ylabel('')  # Remove the y-axis label for a cleaner look
plt.figtext(0.5, -0.1, 'This pie chart shows the breakdown of customers by their subscription status.', ha='center', fontsize=12)
plt.show()
```

#### 8. Most popular payment methods

```python
gb_payment = df.groupby(['Payment Method'])['Customer ID'].count().reset_index(name='Count')
sns.barplot(data=gb_payment, x='Payment Method', y='Count', palette='cool')
plt.title('Most Popular Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

##### This barplot shows that Credit Card is most popular payment method among the customers.

#### 9. Purchase amounts based on promo code usage

```python
promo_groupby = df.groupby('Promo Code Used')['Purchase Amount (USD)'].sum().reset_index()
sns.barplot(data=promo_groupby, x='Promo Code Used', y='Purchase Amount (USD)', palette='muted')
plt.title('Purchase Amount by Promo Code Usage')
plt.xlabel('Promo Code Used')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

#### 10. Frequency of purchases by age groups

```python
age_bins = [0, 25, 35, 45, 55, 65, 100]  # Adjust as needed
age_labels = ['<25', '25-35', '35-45', '45-55', '55-65', '65+']
df['Age_category'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
age_group_purchase = df.groupby('Age_category')['Purchase Amount (USD)'].sum().reset_index()
sns.barplot(data=age_group_purchase, x='Age_category', y='Purchase Amount (USD)', palette='plasma')
plt.title('Purchase Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

#### 11. Correlation between product size and purchase amount

```python
size_group = df.groupby('Size')['Purchase Amount (USD)'].sum().reset_index()
sns.barplot(data=size_group, x='Size', y='Purchase Amount (USD)', palette='coolwarm')
plt.title('Purchase Amount by Product Size')
plt.xlabel('Product Size')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

##### Above Barplot shows the correlation between the size of the product and the purchase amount.

#### 12. Preferred shipping type by product category

```python
shipping_category = df.groupby(['Shipping Type', 'Category']).size().reset_index(name='Count')
sns.barplot(data=shipping_category, x='Shipping Type', y='Count', hue='Category')
plt.title('Preferred Shipping Type by Product Category')
plt.xlabel('Shipping Type')
plt.ylabel('Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### 13. Impact of discount application on purchase amounts

```python
discount_group = df.groupby('Discount Applied')['Purchase Amount (USD)'].sum().reset_index()
sns.barplot(data=discount_group, x='Discount Applied', y='Purchase Amount (USD)', palette='Set2')
plt.title('Purchase Amount by Discount Application')
plt.xlabel('Discount Applied')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

####  16. Purchase amount by review ratings

#### 17. Purchase behavior by location

```python
location_purchase = df.groupby('Location')['Purchase Amount (USD)'].sum().reset_index()
location_purchase = location_purchase.sort_values(by='Purchase Amount (USD)', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=location_purchase, x='Location', y='Purchase Amount (USD)', palette='cubehelix')
plt.title('Total Purchase Amount by Location')
plt.xlabel('Location')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.show()
```
